## Sentiment Anaysis

Hangi task uzerinde calismak istiyorsak pipeline uzerinden bunu bize sagliyor .

__pipeline -->__ Hangi task uzerinde calisacagimizi icindeki __task__ parametresine yazilir. Biz burda sentiment analiz yapacagimiz icin unu yazdik :

__pipeline__ icindeki __model__ parametresine hangi modeli kullanacaksak bunu yazabiliriz. Burda biz yazmadigimiz icin default olan model gelecek.

pipeline' i bir degisken icine task ile birlikte tanimladik ve bu degisken icine bir cumle yazdik. Kodu calistirinca hazir model huggingface sayfasindan cekildi (Ciktida modelin web adresi var. Biz model ismi belirtmedigimiz icin default model geldi). Model once iner, daha sonra calisir (Model, 255 MB imis)

Model, yorumun positive oldugunu %95' lik bir skorla bildi :

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I've been waiting for a HuggingFace course my whole life.")
```

Yukaridaki kodda sadece modele verdigimiz cumleyi degistirerek tekrar calistirdik. Model, bu cumlenin de olumlu oldugunu %99 olasilikla bildi :

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I love you.")
```

Asagida virgulle ayrilmis sekilde olumlu ve olumsuz cumle yazilip modele verilmis. Model, bu cumlelerin de olumlu-olumsuz olma durumunu cok yuksek olasilikla bildi :

```python
classifier(
    ["I've been waiting for a HuggingFace course my whole life.", "I hate this so much!"]
)
```

## Zero-Shot-Classification

__candidate_labels__ parametresi ile class' lari kendimiz belirleriz. Asagidaki class' lara ilaveten baska class' lar da ilave edilebilir, eklenen class ile ilgili mutlaka bir sonuc dondurur ('art', 'sport', 'health' gibi). 

Modeli belirtmedigimiz icin default modeli secti (Ciktida web adresi var). Model, 1.52GB imis.

Model, verilen cumlenin egitimle alakali oldugunu %84 oraninda bilmis :

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)
```

Yukaridaki cumleye ek olarak 2 cumle daha yazdik ve class isimlerini de artirdik. Birden fazla cumle yazilacaksa cumlelerin hepsi mutlaka bir [] icine alinmalidir.

Ciktida cumlelerin sirayla hangi class' tan ne kadar skor aldiklarini goruyoruz.Ornegin son cumlenin medikal ile alakali oldugunu yuksek olasiliklarla bilmis :

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    ["This is a course about the Transformers library", "Barcelona wins Real Madrid", "Doctor works at hospital"],
    candidate_labels=["education", "politics", "business", "sport", "medikal"],
)
```

## text-generation

Cumlenin baslangici verilir, son kismini model kendisi doldurur. Bu durumlarda default olarak __gpt2__ modeli kullanilir.

Asagida yarim birakilan texti model kendisi tamamladi :

```python
from transformers import pipeline

generator = pipeline("text-generation")
generator("In this course, we will teach you how to")
```

Asagida default model yerine model ismini kendimiz sectik. __distilgpt2__, gpt2' nin bir modelidir. Text olarak ayni cumleyi verdik. 

__max_length=30 -->__ Her cumle max 30 token uzunlugunda olsun.

__num_return_sequences=2 -->__ Verilen text ile ilgili 2 farkli cumle doldur.

```python
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")
generator(
    "In this course, we will teach you how to",
    max_length=30,
    num_return_sequences=2,
)
```

## fill-mask

fill-mask ile bir sonraki tokeni tahmin etme islemi yapilir. 

Hangi token tahmin edilmek isteniyorsa o kisma __<mask>__  yazilir.

Default model __distilroberta-base__ model imis.

__top_k=2 -->__ Iki farkli tahmin döndür. (Ilki en fazla olasiligi, ikincisi de en fazla ikinci olasiligi döndürür)

Ilk cumledeki boslugu _mathematical_ ile doldurmus. Model bu tokeni %19 olasilikla bulmus ve bu token corpusta 30412. tokenmis bilgilerini de ciktida veriyor. %19 dusuk bir olasilik gibi gorunse de, bircok token arasindan bu olasiligi döndürdugu icin aslinda yuksek bir skor. 

Ikinci ihtimal olarak _computational_ kelimesini %4 olasilikla bulmus ve bu token corpusta 38163. sirada.

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker("This course will teach you all about <mask> models.", top_k=2)
```

Fill-mask ile modelin I love' dan sonra gelen kelimeyi tahmin etmesini istedik. %3 olasilikla __him__ kelimesini tahmin etti. Bu tokenin corpustaki yeri 123. sira imis. 

Hazir modelleri kullanirken Google, Microsoft gibi buyuk firmalarin yaptiklari modelleri tercih etmek gerekir. 

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker("I love <mask>", top_k=2)
```

## Ner (Name Entity)

Corpusta gecen yer isimleri, kisiler, organizasyonlar, tarihleri bularak bunlari etiketler. Daha cok hastalik tespitleri gibi sagliksal problemlerde tercih edilir. 

Asagidaki verilen ornekte model, Sylvain' in kisi oldugunu (PER) %99 olasilikla bilmis. 

__start 11, end 18 :__ Sylvain, cumlede 11. indexte baslayip 18. indexte bitmis.

ORG : Organization ; LOC : Location

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("My name is Sylvain and I work at Hugging Face in Brooklyn.")
```

## Question-Answering

Bir text verilir ve bu text' e gore bir soru sorulur, modelin bu soruyu bilmesi beklenir. Burada model %69 olasilikla soruyu dogru cevaplamis. 

```python
from transformers import pipeline

question_answerer = pipeline("question-answering")
question_answerer(
    question="Where do I work?",
    context="My name is Sylvain and I work at Hugging Face in Brooklyn",
)
```

Eger birden fazla soru sorulacaksa asagidaki gibi sorular [] icine yazilir :

Model sorulan her soruya dogru cevap verdi :

```python
from transformers import pipeline

question_answerer = pipeline("question-answering")
question_answerer(
    question=["Where do you work?", "In which city do you work?", "What is your name?"],
    context="My name is Sylvain and I work at Hugging Face in Brooklyn",
)
```

## Summarization

Verilen text' in ozetini cikarir. Butun bir cumlenin word embedding' ini alir ve en fazla agirligi olan keyword' ler secer. Bu keyword' ler uzerinden ozet cikarir :

```python
from transformers import pipeline

summarizer = pipeline("summarization")
summarizer(
    """
    America has changed dramatically during recent years. Not only has the number of 
    graduates in traditional engineering disciplines such as mechanical, civil, 
    electrical, chemical, and aeronautical engineering declined, but in most of 
    the premier American universities engineering curricula now concentrate on 
    and encourage largely the study of engineering science. As a result, there 
    are declining offerings in engineering subjects dealing with infrastructure, 
    the environment, and related issues, and greater concentration on high 
    technology subjects, largely supporting increasingly complex scientific 
    developments. While the latter is important, it should not be at the expense 
    of more traditional engineering.

    Rapidly developing economies such as China and India, as well as other 
    industrial countries in Europe and Asia, continue to encourage and advance 
    the teaching of engineering. Both China and India, respectively, graduate 
    six and eight times as many traditional engineers as does the United States. 
    Other industrial countries at minimum maintain their output, while America 
    suffers an increasingly serious decline in the number of engineering graduates 
    and a lack of well-educated engineers.
"""
)
```

## Text Classification

huggingface.com sayfasinda __models__ kismindan Text_Clasiification'i aldik ve pipeline icine tanimladik. 

Model secimi icin tekrar huggingface sayfasina donduk ve Text Classification ile alakali modellerden birini sectik. Modelin sayfasina girince model isminin ustunde __copy__ secenegi ile model ismini kopyalayip asagidaki model kismina yapistirdik. label olarak 0 ciktisini verdi. Bu labelin ne anlama geldigi sayfasinda ayrintili olarak yazar :

```python
from transformers import pipeline
translator = pipeline(task = "text-classification", model="cross-encoder/ms-marco-MiniLM-L-12-v2")
translator("I love you.")
```

Asagida text classification yapan baska bir modeli huggingface sayfasindan aldik. Bu modelde labellar starlar ile ifade edilmis. Sentiment analiz oldugu icin model bu cumleye 5 stars verdi ve olumlu oldugunu bildi :

```python
from transformers import pipeline
translator = pipeline(task = "text-classification", model="nlptown/bert-base-multilingual-uncased-sentiment")
translator("I love you.")
```

Asagidaki gibi task kismina yanlis bir isim yazdigimizda oraya neler yazabilecegimizin listesi ciktida verilmis. Bunlardan herhangi biri task olarak yazilabilir :

```python
from transformers import pipeline

translator = pipeline("bbb", model="Helsinki-NLP/opus-mt-fr-en")
translator("I love you.")
```

### Fill-Mask Ornegi 

__mask__ belirtilirken __[MASK]__ seklinde belirtilmesi gerektigi modelin huggingface sayfasindaki aciklamasinda belirtilmis. 'I love' dan sonra en cok nelerin geldigi ciktida var :

```python
from transformers import pipeline
translator = pipeline(task = "fill-mask", model="bert-base-uncased")
translator("I love [MASK]")
```
